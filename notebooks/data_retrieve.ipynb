```python
import eikon as ek
import pandas as pd
from tqdm import tqdm

eikon_key=''#insert the eikon key here, deleted for security reasons
ek.set_app_key(eikon_key)
```

```python
# Fetch constituents
euro50_constituents, err = ek.get_data('0#.STOXX50E', ['TR.RIC'])
# Convert to sets for easier operations
euro50_tickers = list(euro50_constituents['RIC'])
```

```python
print(f"Number of tickers: {len(euro50_tickers)}")
print(euro50_tickers)
```

```python
# Fetch prices for RICs one by one
df_prices = pd.DataFrame()


# Example loop for fetching data (replace euro50_tickers with your list of tickers)
for ric in tqdm(euro50_tickers):
    df_prices_batch = ek.get_timeseries(
        ric, fields='CLOSE',
        start_date='2023-01-01',
        end_date='2024-12-31',
        interval='daily',
        calendar='tradingdays'
    )
    # Use RICs as column names
    df_prices_batch.columns = [ric]
    # Concatenate the batch data with the existing dataframe
    df_prices = pd.concat([df_prices, df_prices_batch], axis=1, join='outer')

df_prices
```

```python
# save to csv
df_prices.to_csv("../data/euro50_prices.csv")
```

```python
fields = [
    "TR.TRESGScore",                  # Thomson Reuters ESG Score
    "TR.EnvironmentPillarScore",
    "TR.SocialPillarScore",
    "TR.GovernancePillarScore"
]

df_shares, err = ek.get_data(
    euro50_tickers,
    ['TR.CompanyMarketCap.Date' ,'TR.CompanyMarketCap','TR.CompanyMarketCap.Currency','TR.F.ComShrOutsTot'],
    {'SDate': '2023-01-01', 'EDate': '2024-12-31', 'FRQ':'D','period':'FY0'},
)


df_shares
```

```python
df_shares.to_csv("../data/euro50_shares.csv", index=False)
```

```python
# Convert Date column to datetime
df_shares['Date'] = pd.to_datetime(df_shares['Date'])

# Pivot the dataframe
df_common_shares = df_shares.pivot(
    index='Date',
    columns='Instrument',
    values='Common Shares - Outstanding - Total'
)

# Convert date index to proper format 'yyyy-mm-dd'
df_common_shares.index = df_common_shares.index.strftime('%Y-%m-%d')

# Fill NaN values using forward fill method
df_common_shares = df_common_shares.fillna(method='ffill')

# Forward fill remaining NaN values at the start using backward fill
df_common_shares = df_common_shares.fillna(method='bfill')

# Format numbers to remove scientific notation and round to 2 decimal places
df_common_shares = df_common_shares.round(2)

df_common_shares
```

```python
df_marketcap = df_shares.pivot(
    index='Date',
    columns='Instrument',
    values='Company Market Cap'
)

df_marketcap.index = df_marketcap.index.strftime('%Y-%m-%d')
df_marketcap = df_marketcap.fillna(method='ffill')
df_marketcap = df_marketcap.fillna(method='bfill')
df_marketcap = df_marketcap.round(2)


df_marketcap
```
